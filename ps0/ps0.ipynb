```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im1 = cv2.cvtColor(cv2.imread('output/ps0-1-a-1.png'), cv2.COLOR_BGR2RGB)
im2 = cv2.cvtColor(cv2.imread('output/ps0-1-a-2.png'), cv2.COLOR_BGR2RGB)
plt.imshow(im1)
plt.figure()
plt.imshow(im2)
```

```python
# Swap red and blue pixels
im2 = im1.copy()
im2[:,:,0], im2[:,:,2] = im1[:,:,2], im1[:,:,0]
plt.imshow(im2)
cv2.imwrite('output/ps0-2-a-1.png', im2)
```

```python
# Save only green channel
img_green = im1[:,:,1]
plt.imshow(img_green, 'gray')
cv2.imwrite('output/ps0-2-b-1.png',img_green)
```

```python
img_red = im1[:,:,0]
plt.imshow(img_red, 'gray')
cv2.imwrite('output/ps0-2-c-1.png',img_red)
```

```python
# Question 4.a
print('Min(img_green) : %d' % np.amin(img_green))
print('Max(img_green) : %d' % np.amax(img_green))
print('Average(img_green) : %.4f' % np.average(img_green))
print('Standard Deviation (img_green) : %.4f' % np.std(img_green))
```

```python
# Question 4.b
mean = int(np.average(img_green))
img = img_green.copy()
img -= mean
img = img*10
img += mean
plt.imshow(img, 'gray')
cv2.imwrite('output/ps0-4-b-1.png', img)
```

```python
# Question 4.c
img = img_green.copy()
img[:,0:-2] = img[:,2:]
img[:,-2:] = img_green[:,:2]
cv2.imwrite('output/ps0-4-c-1.png', img)
plt.imshow(img, 'gray')
```

```python
# Question 4.d
im1_diff = img_green - img
cv2.imwrite('output/ps0-4-d-1.png', im1_diff)
plt.imshow(im1_diff, 'gray')
```

```python
# Question 5.a
im_blur = cv2.GaussianBlur(im1,(11,11),5)
plt.imshow(im_blur)
cv2.imwrite('output/ps0-5-a-1.png', im_blur)
```

```python
im_blur2 = im1.copy()
im_blur2[:,:,2] = cv2.GaussianBlur(im_blur2[:,:,2],(21,21),5)
plt.imshow(im_blur2)
cv2.imwrite('output/ps0-5-b-1.png', im_blur2)
```
